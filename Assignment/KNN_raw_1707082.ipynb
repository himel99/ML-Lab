```python
test_index = []
rand
train = test_data

for i in range(len(rand)):
    test_index.append(rand.index[i])
print(test_index)
train = train.drop(train.index[test_index])
train
```

# KNN Function

```python
from statistics import mode

def knn(test, user_data_test, k):
    data_arr = []
    dist_arr = []
    res_test = []

    for l in range(len(user_data_test)):
        print('Iteration : ',l)
        for i in range(0,len(train)):
            for j in columns:
                #print(test[j][i])
                elem = train[j][i]
                data_arr.append(elem)
                #print(data_arr)
                dist = ((user_data_test[l][0]-data_arr[0])**2)**0.5
                dist_arr.append(dist)
                data_arr = []
        result=pd.DataFrame(dist_arr)
        result['Glucose'] = train['Glucose']
        result['Diabetes'] = train['Outcome']

        final = result.nsmallest(k,[0])
        res = final['Diabetes']
        #print(res)
        res = mode(res)
        res_test.append(res)
        print('Prtedicted : ',user_data_test[l][1],' Actual : ', res)
        data_arr = []
        dist_arr = []
    print(res_test)  

    total = 0
    yes = 0
    no = 0

    for i in range(len(res_test)):
        total += 1
        if(res_test[i] == user_data_test[i][1]):
            yes +=1
        else: 
            no+=1

    print('Total Data : ', total, ' Correct :', yes,' Incorrect :', no)
    return no
```

# Error Rate Plotting with various K

```python
import matplotlib.pyplot as plt

validation_data = 10  

rand1 = test_data.sample(n = validation_data)
print(rand1)

user_data_test = rand1.to_numpy()
print(user_data_test[0][1])

error = []

# Calculating error for K values between 1 and 40
for i in range(1, 80):
    pred_error = knn(test, user_data_test, i)
    error.append(np.mean(pred_error))
```

```python
plt.figure(figsize = (12,6))
plt.plot(range(1,80), error, color = 'red', linestyle = 'dashed', marker = 'o', markerfacecolor = 'blue', markersize = 10)
plt.title('Error Rate K Value')
plt.xlabel('K value')
plt.ylabel('Mean Error')
```
